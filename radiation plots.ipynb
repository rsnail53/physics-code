```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import statistics
import math
import pandas as pd
import scipy.constants
from matplotlib.ticker import FuncFormatter
from scipy import optimize
```

```python
def halfwave(r,w):
    #<S> for a halfwave antenna
    return ((np.cos((np.pi/2)*np.cos(w))**2/(np.sin(w))**2)/r**2)
def dipole(r,w):
    #<S> for a perfect electric dipole
    return (np.sin(w)**2)/(r**2)
```

```python

fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
R=1
w= np.linspace(0.1,2*np.pi,5000)


 
ax.plot(w, halfwave(R,w),label="R")
ax.plot(w, halfwave(1.2*R,w), label="1.2*R")
ax.plot(w, halfwave(1.4*R,w), label="1.4*R")
ax.plot(w, halfwave(1.6*R,w), label="1.6*R")
ax.plot(w, halfwave(1.8*R,w), label="1.8*R")
ax.plot(w, halfwave(2*R,w), label="2.0*R")
textfit = "<S> for halfwave antenna at different radii"
ax.text(0, 1.2, textfit, transform=ax.transAxes , fontsize=12,verticalalignment='top')

plt.legend(bbox_to_anchor=(1.5, 1))

fig, ay = plt.subplots(subplot_kw={'projection': 'polar'})
ay.plot(w, dipole(R,w)-halfwave(R,w),label="R")
ay.plot(w, dipole(1.2*R,w)-halfwave(1.2*R,w), label="1.2*R")
ay.plot(w, dipole(1.4*R,w)-halfwave(1.4*R,w), label="1.4*R")
ay.plot(w, dipole(1.6*R,w)-halfwave(1.6*R,w), label="1.6*R")
ay.plot(w, dipole(1.8*R,w)-halfwave(1.8*R,w), label="1.8*R")
ay.plot(w, dipole(2.0*R,w)-halfwave(2.0*R,w), label="2.0*R")
textfit = "Difference in radiation between halfwave antenna and electric dipole"
ay.set_title("Difference in radiation between halfwave antenna and electric dipole")
```
